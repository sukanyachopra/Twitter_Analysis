```python
csv_columns=['Date_Time','User','Author','Tweet_text','Likes','Retweets']
with open('microsoft_tweets.csv','w') as file1:
  writer=csv.DictWriter(file1,fieldnames=csv_columns)
  writer.writeheader()
  for i in cursor:
    if(str(i.created_at)>='2022-07-02' and str(i.created_at)<='2022-07-09'):
      t={'Date_Time':i.created_at,'User':i.user.screen_name,'Author':i.author.screen_name,'Tweet_text':i.full_text,'Likes':i.favorite_count,'Retweets':i.retweet_count}
      writer.writerow(t)
```
